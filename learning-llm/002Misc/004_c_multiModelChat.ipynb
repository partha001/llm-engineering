```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)

OLLAMA_BASE_URL = "http://localhost:11434/v1"
llama_a = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
llama_b = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")

llama_a_system_prompt = """
You are a chatbot who is very polite and thoughtful. You try to find common ground and respond with empathy.
"""

llama_b_system_prompt = """
You are a chatbot who is arrogant and argumentative. You always try to uplift the conversation and encourage others.
"""

# Shared conversation history
conversation_history = []

def call_llama(model_instance, system_prompt, user_message):
    messages = [{"role": "system", "content": system_prompt}]
    messages += conversation_history
    messages.append({"role": "user", "content": user_message})

    response = model_instance.chat.completions.create(model="llama3.2", messages=messages)
    reply = response.choices[0].message.content

    conversation_history.append({"role": "user", "content": user_message})
    conversation_history.append({"role": "assistant", "content": reply})

    return reply

# Start the conversation
initial_message = "Hi there! How are you feeling today?"
print("User:", initial_message)

llama_a_response = call_llama(llama_a, llama_a_system_prompt, initial_message)
print("LLaMA A:", llama_a_response)

# Conversation loop
num_turns = 4  # You can change this to any number
for i in range(num_turns):
    llama_b_response = call_llama(llama_b, llama_b_system_prompt, llama_a_response)
    print("LLaMA B:", llama_b_response, end="\n\n")

    llama_a_response = call_llama(llama_a, llama_a_system_prompt, llama_b_response)
    print("LLaMA A:", llama_a_response, end= "\n\n")
```
