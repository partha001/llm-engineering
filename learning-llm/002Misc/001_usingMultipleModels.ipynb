```python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')
```

```python
messages = [{"role":"user", "content": "tell me funny joke related to programming."}]
```

```python
# asking openai to tell me a joke
openai = OpenAI()
# now making the call to openai using the chat-completions api as shown below 
response = openai.chat.completions.create(model="gpt-5-nano", messages= messages)
display(Markdown( response.choices[0].message.content))
```

```python
# asking google gen ai to tell me joke
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

## and the rest of hte code remains same:
response = gemini.chat.completions.create(model="gemini-2.5-pro", messages= messages)
display(Markdown( response.choices[0].message.content))
```

```python
# askign ollama running locally to tell a joke
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url = OLLAMA_BASE_URL, api_key= "ollama")
response = ollama.chat.completions.create(model="llama3.2", messages= messages)
display(Markdown( response.choices[0].message.content))
```
