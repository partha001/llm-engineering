```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
```

```python
# defining our system prompt - one can experiment with this later. changteing 
# the last sentence to 'Respond in markdown in Spanish.'

system_prompt = """
You are an assistant that analyzes the contents of a website,
and provide a short summarry, ignoring text that mgiht be navigation related.
Respond in markdown.
"""
```

```python
# defining our user prompt
user_prompt_prefix = """
Herea re the contents of a website.
Provide a short summary of this website in markdown. 
If it includes news or announcements, then summarize these too.
"""
```

```python
def messages_for(website):
    return [
        {"role": "system", "content":system_prompt},
       { "role":"user", "content": user_prompt_prefix + website}
    ]
```

```python
#messages_for(ed)
```

```python
openai = OpenAI()
response = openai.chat.completions.create(model="gpt-5-nano", messages= messages_for(ed))
print(response.choices[0].message.content)
```
