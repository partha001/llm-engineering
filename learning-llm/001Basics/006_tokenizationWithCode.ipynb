```python
import tiktoken

# Use a known encoding â€” 'cl100k_base' is commonly used for GPT-4 and GPT-3.5
encoding = tiktoken.get_encoding("cl100k_base")

tokens = encoding.encode("Hi my name is Partha")
print(tokens)
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} == {token_text}")
```
