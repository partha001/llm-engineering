### making openai calling using http request

```python
import requests

headers = {"Authorization": f"Bearer {api_key}", "Content-Type":"application/json"}

payload = {
    "model": "gpt-5-nano",
    "messages": [{"role":"user", "content":"Tell me a fun fact"}]
}

response = requests.post(
    "https://api.openai.com/v1/chat/completions",
    headers = headers,
    json = payload
)

# thus the complete response payload
response.json()
```

```python
# note that the response object has a lot of other metadata which we will explore later . however only the response text can be accessed as 
response.json()["choices"][0]["message"]["content"]
```

### thus here we are doing the same thing i.e. calling openai but instead of using the python openai client library we are making http call

it is also to be noted that since parsing the json response can be cumbersome and tidious hence using the python client library really makes life easy. actually the openai python client library makes the same http call behind the screen.
