```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
## first visit http://localhost:11434 . if it says running then the prerequisite steps are not required. else do the prerequisite steps

# #prerequisite do the following first:
# step1. execute the command:           ollama pull llama3.2
# step2. run the command :  ollama serve
# step3. verify by visiting http://localhost:11434
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url = OLLAMA_BASE_URL, api_key= "ollama")
messages = [{"role": "user" , "content": "tell me a fun fact"}]

response = ollama.chat.completions.create(model="llama3.2", messages= messages)
response.choices[0].message.content
```
