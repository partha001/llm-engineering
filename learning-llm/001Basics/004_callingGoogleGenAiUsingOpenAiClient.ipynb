### this notebook shows the openai python client library makes like easy by using a common interface to interact with different ai models

```python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
## earlier we have called the openai using the python openai client as shown below:
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
messages = [{"role": "user" , "content": "tell me a fun fact"}]
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages= messages)
response.choices[0].message.content
```
